```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

# Import Dataset

```python
cases= pd.read_csv('DatasetAfricaMalaria.csv')
cases.head()
```

# Visualization for more insights

```python
df_corr_mal = cases[['Malaria cases reported',
'People using safely managed drinking water services, rural (% of rural population)',
'People using safely managed drinking water services, urban (% of urban population)',
'People using safely managed sanitation services, rural (% of rural population)',
'People using at least basic drinking water services, urban (% of urban population)',
'People using at least basic drinking water services, rural (% of rural population)',
'People using at least basic drinking water services, urban (% of urban population)',
'People using at least basic sanitation services, rural (% of rural population)',
'People using at least basic sanitation services, urban  (% of urban population)'
]]

# visualize the correlation
plt.figure(figsize=(15,10))
sns.heatmap(df_corr_mal.iloc[:,0:15].corr(), annot=True,fmt=".0%")
plt.show()
```

```python
df_corr_pop = cases[['Malaria cases reported','Rural population (% of total population)',
       'Rural population growth (annual %)',
       'Urban population (% of total population)',
       'Urban population growth (annual %)'
]]

# visualize the correlation
plt.figure(figsize=(15,10))
sns.heatmap(df_corr_pop.iloc[:,0:15].corr(), annot=True,fmt=".0%")
plt.show()
```

```python
df_corr_treat = cases[['Malaria cases reported','Use of insecticide-treated bed nets (% of under-5 population)',
       'Children with fever receiving antimalarial drugs (% of children under age 5 with fever)',
       'Intermittent preventive treatment (IPT) of malaria in pregnancy (% of pregnant women)'
]]

# visualize the correlation
plt.figure(figsize=(15,10))
sns.heatmap(df_corr_treat.iloc[:,0:15].corr(), annot=True,fmt=".0%")
plt.show()
```

# Top 50 countries with largest number of cases in 2017

```python
df1 = cases.sort_values(by = 'Malaria cases reported', ascending=False)
df1 = df1[df1['Year']==2017][:50]
plt.figure(figsize=(18,10))
sns.barplot(df1['Country Name'], df1['Malaria cases reported'])
plt.xticks(rotation=90, ha='right');
```

### Relationship between number of Malaria cases reported and Incidence of malaria (per 1,000 population at risk)

```python

plt.figure(figsize =(8,5))
plt.title('Relationship between number of Malaria cases reported and Incidence of malaria (per 1,000 population at risk)',
          fontsize=14, fontweight='bold')
ax =sns.scatterplot(x= 'Malaria cases reported', y ='Incidence of malaria (per 1,000 population at risk)'
                    ,data = cases)
```

```python
num_per_year=cases.groupby('Year').sum()
#print(np.sort(num_per_year['Malaria cases reported']).tolist())
num_per_year['Malaria cases reported']
```

```python
df_mal =cases.groupby('Year').sum().loc[:, ['Malaria cases reported']]
plt.figure(figsize =(8,5))
ax = sns.lineplot(data=df_mal,markers=True)
plt.xlabel('Year', fontsize=15)
plt.title('Malaria cases reported by year', fontsize=14, fontweight='bold',color='red' );
plt.show();
```

### Malaria cases reported and Incidence of malaria (per 1,000 population at risk) per Country

```python
df_mal_code =cases.groupby('Country Code').sum().loc[:, ['Malaria cases reported',
                                                         'Incidence of malaria (per 1,000 population at risk)']]

print(df_mal_code.head())


plt.figure(figsize =(30,7))
ax = sns.lineplot(data=df_mal_code,markers=True)
plt.xlabel('Year', fontsize=15)
plt.title('Malaria cases reported and Incidence of malaria (per 1,000 population at risk) in particular Country', 
          fontsize=20, fontweight='bold')
plt.show()
```
